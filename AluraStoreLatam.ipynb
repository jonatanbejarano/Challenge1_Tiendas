### ¿Qué tienda vender?


**Introducción**

Se busca determinar cuál de las cuatro tiendas de Alura Store debería vender el señor Juan. Para ello, se realizará un análisis de datos que abarcará ventas, rendimiento y reseñas de cada tienda. El objetivo es identificar la tienda más conveniente y recomendar la mejor opción para optimizar sus resultados comerciales.
  

**Justificación**

Luego de realizar el respectivo análisis de los datos obtenidos de cada una de las tiendas del señor Juan, se determinaron los resultados correspondientes a los ingresos, costos de envío, promedios de calificación y ganancias netas.

**Ingresos por tienda**

Cada tienda presenta un nivel de ingresos distinto, reflejando variaciones en su rendimiento comercial. Los ingresos totales son los siguientes:

Tienda 1: $1.150.880.400

Tienda 2: $1.116.343.500

Tienda 3: $1.098.019.600

Tienda 4: $1.038.375.700

**Gastos por envío**

En cuanto a los costos asociados al envío de los productos, se observan los siguientes valores por tienda:

Tienda 1: $61.377.900

Tienda 2: $59.485.100

Tienda 3: $58.516.600

Tienda 4: $55.317.400

**Promedio de calificación**

El nivel de satisfacción de los clientes, medido a través del promedio de calificación de cada tienda, muestra resultados positivos y bastante equilibrados:

Tienda 1: 3.98

Tienda 2: 4.04

Tienda 3: 4.05

Tienda 4: 4.00

**Ganancias netas**

Las ganancias netas se calcularon restando los costos de envío a los ingresos totales de cada tienda. De esta manera, se obtuvieron los siguientes resultados:

Tienda 1: $1.089.502.500

Tienda 2: $1.056.858.400

Tienda 3: $1.039.503.000

Tienda 4: $983.058.300

**Conclusión**

De acuerdo con los resultados, se evidencia que la Tienda 1 es la que genera mayores ingresos y mayores ganancias netas, mientras que la Tienda 3 presenta la mejor calificación promedio por parte de los clientes, lo que refleja una buena percepción del servicio. Sin embargo, al comparar los valores financieros globales, la Tienda 4 muestra los ingresos y ganancias netas más bajas, a pesar de tener una calificación aceptable.

Por lo tanto, la Tienda 4 es la que debería el señor juan considerar vender, ya que es la menos rentable dentro del grupo analizado y presenta un menor margen de ganancia respecto a las demás tiendas. Mantener las tres tiendas restantes sería una decisión más favorable para conservar la estabilidad económica y aprovechar el buen desempeño tanto financiero como de satisfacción del cliente.
